```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread('/Users/xuhuahu/Downloads/640x480-test/image-0.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
imgR = img[:,:,0]
imgG = img[:,:,1]
imgB = img[:,:,2]
_, imgR_th = cv2.threshold(imgR, 90, 255, cv2.THRESH_BINARY_INV)
_, imgG_th = cv2.threshold(imgG, 70, 255, cv2.THRESH_BINARY_INV)
_, imgB_th = cv2.threshold(imgB, 60, 255, cv2.THRESH_BINARY_INV)
img_th = cv2.bitwise_and(imgR_th,imgG_th)
img_th = cv2.bitwise_and(img_th,imgB_th)
img_th = cv2.dilate(img_th,kernel=None)
plt.imshow(img_th, cmap='gray', interpolation = 'bicubic',shape=(640,480))
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
_, contours, _ = cv2.findContours(img_th, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
for c in contours:
    r = cv2.boundingRect(c)
    cv2.rectangle(img, (r[0],r[1]), (r[0]+r[2],r[1]+r[3]), (255,255,0), 2)
plt.imshow(img,interpolation = 'bicubic',shape=(640,480))
plt.xticks([]), plt.yticks([])
plt.show()
```
